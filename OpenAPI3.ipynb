```python
import requests
import pprint
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
df = pd.DataFrame()
```

### API 받아오기

```python
start_num = 1
end_num = start_num + 999
while True:

    url = 'http://openapi.foodsafetykorea.go.kr/api/54ec31dae60f46ceb2b8/I2500/json/{0}/{1}/INDUTY_CD_NM=일반음식점'.format(start_num, end_num)                                                                                                        
    response = requests.get(url)
    dict = json.loads(response.text)
    doc = dict["I2500"]["row"]
    df = pd.concat([df, pd.DataFrame(doc)], axis = 0, ignore_index = True)
    start_num += 1000
    
    end_num = start_num + 999
    print(url)
```
