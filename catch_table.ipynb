```python
def crawling():
    scroll_location = browser.execute_script("return window.pageYOffset")
    restaurants, rates, reviews = [], [], []
    a = 0
    while True:
        restaurant = "#searchList__{} > div > div > div > div > div.detail > div.detail-header > div > h4".format(a)
        rate = "#searchList__{} > div > div > div > div > div.detail > div.restaurant-meta > div._1xnbt310._1xnbt312 > div > span._1xnbt316._1ltqxco1b".format(a)
        review = "#searchList__{} > div > div > div > div > div.detail > div.restaurant-meta > div._1xnbt310._1xnbt312 > div > span._1xnbt317._1ltqxco1e".format(a)

        if len(browser.find_elements(By.CSS_SELECTOR, rate)) > 0:
            for i in browser.find_elements(By.CSS_SELECTOR, restaurant):
                restaurants.append(i.text)
                print(i.text)
    
            for i in browser.find_elements(By.CSS_SELECTOR, rate):
                rates.append(i.text)
                print(i.text)
    
            for i in browser.find_elements(By.CSS_SELECTOR, review):
                reviews.append(i.text[1:-1])
                print(i.text[1:-1])

        browser.execute_script("window.scrollTo(0,{})".format(scroll_location + 230))
        time.sleep(1)
        scroll_height = browser.execute_script("return window.pageYOffset")
        
        if scroll_location == scroll_height:     # 스크롤이 끝까지 내려가면 종료
            break
            
        else:
            scroll_location = browser.execute_script("return window.pageYOffset")     
       
        a += 1

    return restaurants, rates, reviews

restaurants, rates, reviews = crawling()

print("가져온 식당 수: ", len(restaurants))
for r1, r2, r3 in zip(restaurants, rates, reviews):
    print("식당: ", r1)
    print("평점: ", r2)
    print("리뷰 수: ", r3)
    print()
```
