```python
import requests
import pprint
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

### API 받아오기

```python
pageno = 1
#nrow = 10000
year = 2018
df = pd.DataFrame()
```

```python
for i in range(0, 5):
    pageno = 1
    while pageno <= 18:
        url = 'http://apis.data.go.kr/1130000/FftcBrandFrcsUnitAvrSalInfoService/getBrandFrcsUnitAvrSalInfo?serviceKey=Xa9F%2BlohBnyap8Pj%2FHmyDdz%2FozVb5GFsC5z06uWip4HPt7n2yNTwF8UqP3Oq%2F%2Fx2M78iBDteaqxGceksQRSxdA%3D%3D&pageNo={0}&numOfRows=10000&resultType=json&yr={1}'.format(pageno, year)
        response = requests.get(url)
        dict = json.loads(response.text)
        doc = dict["items"]
        df = pd.concat([df, pd.DataFrame(doc)], axis = 0, ignore_index = True)

        print(url)
        pageno += 1
        
    year += 1
```
