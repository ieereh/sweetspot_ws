```python
import requests
import pprint
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
pageno = 1
nrow = 100
year = 2018
td = pd.DataFrame()
```

### API 받아오기

```python
for i in range(0, 5):
    pageno = 1
    while pageno <= 78:
        url = 'http://apis.data.go.kr/1130000/FftcBrandFrcsStatsService/getBrandFrcsStats?serviceKey=Xa9F%2BlohBnyap8Pj%2FHmyDdz%2FozVb5GFsC5z06uWip4HPt7n2yNTwF8UqP3Oq%2F%2Fx2M78iBDteaqxGceksQRSxdA%3D%3D&pageNo={0}&numOfRows={1}&resultType=json&yr={2}'.format(pageno, nrow, year)
        response = requests.get(url)
        dict = json.loads(response.text)
        doc = dict["items"]
        td = pd.concat([td, pd.DataFrame(doc)], axis = 0, ignore_index = True)

        print(url)
        pageno += 1
        
    year += 1
```
