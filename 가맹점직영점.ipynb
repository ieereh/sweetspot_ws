```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
```

```python
chrome_options = Options() # 브라우저가 자꾸 꺼져서 생기는 에러를 방지하기 위해 옵션 추가.. 
chrome_options.add_experimental_option("detach", True)
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])
service = Service(executable_path=r'C:/Users/Puser/Workspace/chromedriver.exe') # 크롬드라이버 위치를 절대경로로 설정
```

```python
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
b_list ,s_list, cate_list, tot_list, f_list, d_list = [], [], [], [], [], []
    
for i in range(35):
    j = 1
    while j <= 300:
        try:
            driver.find_element(By.XPATH, "//*[@id='frm']/table/tbody/tr[{0}]/td[2]/a".format(j)).click()
                
            b_name = driver.find_element(By.XPATH, "//*[@id='frm']/div[2]/div/table[1]/tbody/tr[1]/td[1]").text[3:]
            s_name = driver.find_element(By.XPATH, "//*[@id='frm']/div[2]/div/table[1]/tbody/tr[1]/td[2]").text[5:]
            category = driver.find_element(By.XPATH, "//*[@id='frm']/div[2]/div/table[1]/tbody/tr[1]/td[4]").text
            total = driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div/table[2]/tbody/tr[1]/td[2]").text
            fstore = driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div/table[2]/tbody/tr[1]/td[3]").text
            dstore = driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div/table[2]/tbody/tr[1]/td[4]").text

        except NoSuchElementException as e:
            print("요소를 찾지 못했습니다.", e)
    
        print("상호: {0}\n영업표지: {1}\n업종: {2}\n전체: {3}\n가맹점: {4}\n직영점: {5}".format(b_name, s_name, category, total, fstore, dstore))
        print("-" * 30, f'{i+1} 페이지 {j} 번째') 
            
        b_list.append(b_name)
        s_list.append(s_name)
        cate_list.append(category)
        tot_list.append(total)
        f_list.append(fstore)
        d_list.append(dstore)
            
        driver.back()
        driver.refresh()     # 뒤로가기 실행 시 페이지가 로드되지 않아 새로고침 실행
        time.sleep(2)
            
        j += 1

    try:
        driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div/ul/li[{0}]/a".format(i+4)).click()

    except NoSuchElementException as e:
        print("요소를 찾지 못했습니다.", e)
            
    time.sleep(2)
```

```python
df = pd.DataFrame(b_list, columns=['상호'])

df['영업표지'] = s_list
df['업종'] = cate_list
df['전체'] = tot_list
df['가맹점'] = f_list
df['직영점'] = d_list
```
